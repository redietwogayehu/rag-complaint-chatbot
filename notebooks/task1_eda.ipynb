```python
import pandas as pd

df = pd.read_csv(
    "../data/raw/complaints.csv"
)

print(df.shape)

df.head()
```

```python
df.columns.tolist()
```

```python
df["Product"].value_counts()
```

Filter Products

```python
target_products = [
    "Credit card",
    "Payday loan, title loan, personal loan, or advance loan",
    "Checking or savings account",
    "Money transfer, virtual currency, or money service"
]

filtered_df = df[
    df["Product"].isin(target_products)
].copy()

print(filtered_df.shape)
print(filtered_df["Product"].value_counts())
```

```python
filtered_df = filtered_df[
    filtered_df["Consumer complaint narrative"].notna()
].copy()

print(filtered_df.shape)
```

```python
filtered_df["Product"].value_counts()
```

Create Product Category Column

```python
product_mapping = {
    "Credit card": "Credit Card",

    "Payday loan, title loan, personal loan, or advance loan":
        "Personal Loan",

    "Checking or savings account":
        "Savings Account",

    "Money transfer, virtual currency, or money service":
        "Money Transfer"
}

filtered_df["product_category"] = (
    filtered_df["Product"]
    .map(product_mapping)
)
```

```python
filtered_df["product_category"].value_counts()
```

Clean Complaint Narratives

```python
import re

def clean_text(text):

    text = str(text).lower()

    text = re.sub(r"http\S+", "", text)

    text = re.sub(
        r"[^a-zA-Z0-9\s]",
        " ",
        text
    )

    text = re.sub(
        r"\s+",
        " ",
        text
    )

    return text.strip()
```

```python
filtered_df["clean_text"] = (
    filtered_df["Consumer complaint narrative"]
    .apply(clean_text)
)
```

```python
filtered_df[
    [
        "Consumer complaint narrative",
        "clean_text"
    ]
].head()
```
